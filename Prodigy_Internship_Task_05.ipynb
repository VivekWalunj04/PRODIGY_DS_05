```python
dummy = pd.get_dummies(df2[["Age_band_of_driver","Vehicle_driver_relation","Weather_conditions","Driving_experience","Area_accident_occured","Lanes_or_Medians",
                            "Types_of_Junction","Road_surface_type","Light_conditions","Type_of_collision","Vehicle_movement","Casualty_class","Age_band_of_casualty","Cause_of_accident"]],drop_first=True)
```

```python
dummy.head()
```

```python
df3 = pd.concat([df2,dummy],axis=1)
df3.head()
```

```python
df3.drop(["Age_band_of_driver","Vehicle_driver_relation","Weather_conditions","Driving_experience","Area_accident_occured","Lanes_or_Medians",
                            "Types_of_Junction","Road_surface_type","Light_conditions","Type_of_collision","Vehicle_movement","Casualty_class","Age_band_of_casualty","Cause_of_accident"],axis=1,inplace=True)
```

```python
df3.head()
```

```python
x=df3.drop("Accident_severity",axis=1)
x.shape
```

```python
x.head()
```

```python
y=df3.iloc[:,2]
y.head()
```

```python
y.value_counts()
```

```python
sns.countplot(x=y,palette="plasma")
plt.title("Accident Severity Count")
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_smote,y_smote = smote.fit_resample(x,y)
```

```python
y1=pd.DataFrame(y_smote)
y1.value_counts()
```

```python
sns.countplot(x=y_smote,palette="plasma")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_smote,y_smote,test_size=0.3,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)
```

```python
y_pred=model_KNN.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
report_KNN = classification_report(y_test,y_pred)
print(report_KNN)
```

```python
matrix_KNN = confusion_matrix(y_test,y_pred)
matrix_KNN
```
